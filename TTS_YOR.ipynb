```python
from transformers import VitsModel, AutoTokenizer
import torch

model = VitsModel.from_pretrained("facebook/mms-tts-yor")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-yor")

text = '''Bí olùkọ́ kan fúnra rẹ̀ kò bá mọ̀ pé ẹ̀kọ́ ṣe kókó,\
        kò ní lè jẹ́ olùkọ́ tó múná dóko, kò ní lè jẹ́ olùkọ́ tó ń ṣàṣeyọrí,\
        tó ń mú káwọn èèyàn fẹ́ láti kẹ́kọ̀ọ́ kó sì jẹ́ olùkọ́ tó ní ìtẹ́lọ́rùn.
'''
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform
```

```python
import scipy

scipy.io.wavfile.write("techno.wav", rate=model.config.sampling_rate, data=output)
```
